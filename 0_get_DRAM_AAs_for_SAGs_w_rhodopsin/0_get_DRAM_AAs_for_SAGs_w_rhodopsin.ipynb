# 0. Get DRAM AAs for SAGs with rhodopsin

For rhodopsin analysis, we used DRAM to predict proteins, CDS, and annotations.<Br>
Since 926 out of the ~9000 GORG-Dark SAGs have rhodopsins, we do not care about the other 90% of SAGs.<Br>
<u>The bash scripts here collect the proteins CDS, and annots for those 927 SAGs, catting each together</u> into:<br>
* dram_926_SAGs_combined.fna (CDS)
* dram_926_SAGs_combined.faa (proteins)
* dram_926_SAGs_combined.csv (annotations)
